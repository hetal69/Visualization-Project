```python
# Select the relevant columns
selected_columns = ['Previous qualification (grade)', 'Admission grade']
selected_data = df[selected_columns]

# Remove non-numeric rows from the selected columns
selected_data = selected_data.apply(pd.to_numeric, errors='coerce').dropna()

# Compute statistical measures
statistics = selected_data.describe()
```

```python
# Compute additional statistical measures
statistics.loc['range'] = statistics.loc['max'] - statistics.loc['min']
statistics.loc['IQR'] = statistics.loc['75%'] - statistics.loc['25%']
statistics.loc['mode'] = selected_data.mode().iloc[0]

# Calculate mean and median
statistics.loc['mean'] = selected_data.mean()
statistics.loc['median'] = selected_data.median()

# Calculate sample variance and sample standard deviation
statistics.loc['sample_variance'] = selected_data.var()
statistics.loc['sample_std_dev'] = selected_data.std()
```

```python
# Display statistical measures
print(statistics)
```

```python
# Identify outliers using IQR method
Q1 = statistics.loc['25%', 'Previous qualification (grade)']
Q3 = statistics.loc['75%', 'Previous qualification (grade)']
IQR = Q3 - Q1
outlier_threshold = 1.5 * IQR
outliers = selected_data[selected_data['Previous qualification (grade)'] < Q1 - outlier_threshold]
outliers = outliers.append(selected_data[selected_data['Previous qualification (grade)'] > Q3 + outlier_threshold])
```

```python
# Display outliers
print("Outliers:", outliers)
```

```python
# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=selected_data)
plt.title('Box Plot of Previous qualification (grade) and Admission grade')
plt.ylabel('Grade')
plt.show()
```

```python
# Create Sea-born Simplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=selected_data, x='Previous qualification (grade)', y='Admission grade')
plt.title('Scatter Plot of Previous qualification (grade) vs Admission grade')
plt.xlabel('Previous qualification (grade)')
plt.ylabel('Admission grade')
plt.show()
```

```python
# Create Linear Regression Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=selected_data, x='Previous qualification (grade)', y='Admission grade')
plt.title('Linear Regression Plot of Previous qualification (grade) vs Admission grade')
plt.xlabel('Previous qualification (grade)')
plt.ylabel('Admission grade')
plt.show()
```

```python
# Create HeatMap
plt.figure(figsize=(10, 6))
sns.heatmap(selected_data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('HeatMap of Correlation between Previous qualification (grade) and Admission grade')
plt.show()
```
